```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import dtreeviz
```

```python
data = pd.read_csv(".././data/feature_data.csv")
data.head()
```

```python
data = data.set_index("time")
```

```python
train, test = train_test_split(data, test_size=0.1, shuffle=False)
train.shape, test.shape
```

##### Random Forest Regressor

- Uses multiple decision trees ( weak learners ) for form a strong learner
- Uses Bagging ( bootstrap aggregation ) for prediction

```python
rf_regressor = RandomForestRegressor(
    n_estimators=100, criterion="absolute_error", max_depth=5, min_samples_split=10,
    warm_start=True
)

target_col = "bld_engcons"
x_cols = list(set(train.columns) - set([target_col]))
rf_regressor.fit(train[x_cols].values, train[target_col].values)
rf_regressor
```

```python
viz_rmodel = dtreeviz.model(
    model=rf_regressor, 
    tree_index=1,
    X_train=train[x_cols], 
    y_train=train[target_col],
    feature_names=x_cols, 
    target_name=target_col)
viz_view = viz_rmodel.view()
```
