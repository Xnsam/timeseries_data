```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data.shape
```

```python
data.head()
```

```python
# make column names more code friendly
new_col_names_map =  {i: i.lower().replace(" ", "_") for i in data.columns}

data = data.rename(columns=new_col_names_map)
data.head()
```

```python
data = data.set_index("time")
```

```python
data.head()
```

```python
load_cols = ["comms_and_services", "space_heating", "hot_water", "sockets",
             "lighting", "bld_engcons", "car_chargers", "forecast_uvindex", "forecast_precipitationprobability"]
for col in load_cols:
  print(data[col].value_counts(normalize=True).head(5))
```

```python
data.columns
```

```python
test_data = data.copy()

cols = ["forecast_temperature", "forecast_feelslike"]
test_data = test_data[cols]
test_data = test_data.dropna(subset=cols, how="any")
```

```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
ztest(test_data[cols[0]], test_data[cols[1]], 
      value=data[cols[0]].mean() - data[cols[1]].mean(),
     alternative="two-sided",
     usevar="unequal")
```
